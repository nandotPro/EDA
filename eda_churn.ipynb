# Transform Data

```python
# Transforf 'TotalCharges' from object to float
df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'], errors='coerce')
```

```python
# Show DataFrame schema
df_contracts.info()
```

After transformation, 'TotalCharges' now have 11 rows with missing values.

## Renaming columns

```python
# Renaming with .rename and dict
df_customers.rename(columns={'SeniorCitizen': 'Above65yo'}, inplace=True)
df_customers.info()
```

```python
# Renaming with .columns and list
df_customers.columns = ['IDCliente', 'Genero', 'Mais65Anos', 'TemParceiro', 'TemDependentes']
df_customers.info()
```

## Unifying DataFrames

```python
len(df_customers)
len(df_contracts)
len(df_services)
```

```python
df_services.rename(columns={'customerID': 'IDCliente'}, inplace=True)
df_services.info()
```

```python
# Unifying customer and services DataFrames
df_temp = df_customers.merge(df_services, on='IDCliente')
df_temp.info()
```

```python
# Unifynt def_temp and df_contracts DataFrames
df_churn_temp = df_temp.merge(df_contracts, left_on=['IDCliente'], right_on=['customerID'])
df_churn_temp.info()
```

```python
# unifying all Dataframes with diferent names columns
df_churn = df_customers.merge(df_services, on='IDCliente').merge(df_contracts, left_on=['IDCliente'], right_on=['customerID'])
df_churn.info()
```

```python
# Dropping columns
df_churn.drop(['customerID'], axis=1, inplace=True)
df_churn.info()
```
